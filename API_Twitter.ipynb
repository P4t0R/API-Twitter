```python
#importamos librer√≠as necesarias
import time
import tweepy
from tweepy import OAuthHandler
```

#### **Credenciales**

```python
#Keys
API_KEY = 'xxxxx'
API_SECRET = 'xxxxx'
ACCESS_TOKEN = 'xxxxx-xxxxx'
ACCESS_SECRET = 'xxxxx'
 
auth = tweepy.OAuthHandler(API_KEY, API_SECRET)
auth.set_access_token(ACCESS_TOKEN, ACCESS_SECRET)
 
api = tweepy.API(auth)
```

#### **Leer**

```python
#Usuario a leer
user='elonmusk'

#√öltimo Tweet
tweet = api.user_timeline(id = user, count = 1)[0]
print(tweet.text)
```

```python
#Capturar Tweets
tweets = api.user_timeline(user,count=5)

for tweet in tweets:
  print(tweet.text)
```

```python
#Capturar Tweets
tweets = api.user_timeline(user,count=5,tweet_mode='extended')

for tweet in tweets:
  print(tweet.full_text)
```

```python
#Capturar Tweets con link
def ObtenerTweetsDeUsuario(user_name, number_of_tweets):
    tweets = api.user_timeline(screen_name=user_name, count=number_of_tweets)
    return tweets
 
cnt=0
for tweet in ObtenerTweetsDeUsuario('elonmusk', 5):
  cnt+=1
  print(f'{cnt}.- {tweet.text}')
  print('https://twitter.com/twitter/statuses/' + str(tweet.id))
  print('*'*100)
```
